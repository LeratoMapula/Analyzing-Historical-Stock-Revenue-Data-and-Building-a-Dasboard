```python
import matplotlib.pyplot as plt

def make_graph(data, title):
    plt.figure(figsize=(10, 6)) # Set the size of the graph
    plt.plot(data['Date'], data['Close'], label='Close Price') # Plot the "Close" price
    plt.xlabel('Date') # Label for X-axis
    plt.ylabel('Price (USD)') # Label for Y-axis
    plt.title(title) # Title for the graph
    plt.legend() # Add a legend
    plt.grid() # Add grid lines for better readability
    plt.show() # Display the graph
```

```python
make_graph(tesla_data, "Tesla Stock Closing Prices Over Time")
```

```python
make_graph(gme_data, "GameStop Stock Closing Prices Over Time")
```
