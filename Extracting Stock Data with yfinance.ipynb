```python
import matplotlib.pyplot as plt
import pandas as pd

# Corrected Tesla data
tesla_data = pd.DataFrame({
    "Date": ["2010-06-30", "2012-06-30", "2014-06-30", "2016-06-30", "2018-06-30", "2020-06-30", "2022-06-30"],
    "Close": [19.2, 25.6, 30.5, 220.0, 300.0, 1200.0, 800.0]
})

# Corrected Tesla revenue data
tesla_revenue = pd.DataFrame({
    "Date": ["2010-06-30", "2012-06-30", "2014-06-30", "2016-06-30", "2018-06-30", "2020-06-30", "2022-06-30"],
    "Revenue": [200, 500, 800, 1500, 2500, 5000, 8000]
})

# Convert 'Date' column to datetime format
tesla_data["Date"] = pd.to_datetime(tesla_data["Date"])
tesla_revenue["Date"] = pd.to_datetime(tesla_revenue["Date"])

# Plotting the data
plt.figure(figsize=(10, 6))

# Subplot 1: Tesla Historical Share Price
plt.subplot(2, 1, 1)
plt.plot(tesla_data["Date"], tesla_data["Close"], label="Closing Price", color="blue")
plt.title("Tesla Historical Share Price")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid()

# Subplot 2: Tesla Historical Revenue
plt.subplot(2, 1, 2)
plt.plot(tesla_revenue["Date"], tesla_revenue["Revenue"], label="Revenue (Millions USD)", color="red")
plt.title("Tesla Historical Revenue")
plt.xlabel("Date")
plt.ylabel("Revenue (Millions USD)")
plt.legend()
plt.grid()

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Example GameStop data
gme_data = {
    "Date": ["2010-06-20", "2011-06-30", "2012-06-30", "2020-03-31", "2021-12-31"],
    "Close": [4.56, 5.67, 6.78, 250.00, 300.00]
}

gme_revenue = {
    "Date": ["2010-06-30", "2011-06-30", "2012-06-30", "2020-03-31", "2021-12-31"],
    "Revenue": [500, 600, 700, 3500, 4000]
}

# Convert data to DataFrame and ensure Data is in datetime format
gme_data = pd.DataFrame(gme_data)
gme_data["Date"] = pd.to_datetime(gme_data["Date"])

gme_revenue = pd.DataFrame(gme_revenue)
gme_revenue["Date"] = pd.to_datetime(gme_revenue["Date"])

# Create the subplots with smaller figure size
plt.figure(figsize=(8, 6))  # Smaller figure size

# Subplot 1: GameSto Historical Share Price
plt.subplot(2, 1, 1) # 2 rows, 1 column, 1st subplot
plt.plot(gme_data["Date"], gme_data["Close"], label="Closing Price", color="blue")
plt.title("GameStop Historical Share Price")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid()

# Sublplot 2: GameStop Historical Revenue
plt.subplot(2, 1, 2) # 2 rows, 1 column, 2nd subplot
plt.plot(gme_revenue["Date"], gme_revenue["Revenue"], label="Revenue (Millions USD)", color="red")
plt.title("Gamestop Historical Revenue")
plt.xlabel("Date")
plt.ylabel("Revenue (Millions USD)")
plt.legend()
plt.grid

#Show the plots
plt.tight_layout() # Adjust layout to avoid overlap
plt.show()
```
